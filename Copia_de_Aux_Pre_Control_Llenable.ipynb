```python
def espejoo(p):
  if isinstance(p,Nodoe):
    return p
  else:
    izq = espejoo(p.izq)
    der = espejoo(p.der)
    p.izq = der
    p.der = izq
    return p
```

```python
def espejo(p):
  if p is None or p is Nodoe:
    return None
  else:
    izq = espejoo(p.izq)
    der = espejoo(p.der)
    p.izq = der
    p.der = izq
    return p
```
